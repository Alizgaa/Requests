```python
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
import requests
import lxml

```

```python
url = 'https://spa.gov.mn/newcompanies?/'

ua = UserAgent()
fake_ua = {'user-agent': ua.random}

response = requests.get(url=url, headers=fake_ua)

response.encoding = 'utf-8'

soup = BeautifulSoup(response.text, 'lxml')

table = soup.find('table', class_='table table-striped table-bordered b-t text-small')
```

```python
headers = []

for i in table.find_all('th'):
    title = i.text
    headers.append(title)

#for i in ["Улсын бүртгэлийн дугаар","Үйл ажиллагааны чиглэл","Утас","Факсын дугаар","Хаяг"]:
#    headers.append(i)
headers
```

```python
data = pd.DataFrame(columns=headers)
data
```

```python
div_link_list = [f'https://spa.gov.mn/newcompanies?page={i}' for i in range(1,75)]
div_link_list
```

```python
for lnk in div_link_list:
    url_link = lnk
    response1 = requests.get(url=url_link, headers=fake_ua)
    response1.encoding = 'utf-8'
    
    soup1 = BeautifulSoup(response1.text, 'lxml')
    table1 = soup1.find('table', class_='table table-striped table-bordered b-t text-small')
    
    table1 = table1.find('tbody')

    for i in table1.find_all('tr'):
        row_data = i.find_all('td')
        row = [i.text.strip() for i in row_data]
        length = len(data)
        data.loc[length] = row
print('DONE!')
```
