```python
def crawler(url):
    response = requests.get(url=url, headers=fake_ua)
    response.encoding = 'utf-8'

    soup = BeautifulSoup(response.text, 'lxml')

    table = soup.find('table', class_='table sortable')
    table = table.find('tbody')
    for i in  table.find_all('tr'):
        row_list = i.find_all('td')

        row = [txt.text.strip() for txt in row_list]
        length = len(data)
        data.loc[length] = row
```

```python
for url in ['https://www.tender.gov.mn/mn/result/?year=allYear&tenderCode=undefined&page=1&perpage=10&sortField=&sortOrder=&get=1']:
    try:
        crawler(url)
    except Exception as e:
        pass
```
