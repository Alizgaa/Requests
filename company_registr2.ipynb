```python
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
import requests
import lxml
import time
```

```python
headers = ['Аж ахуйн нэгжийн нэр', 'Регистерийн дугаар', 'Улсын бүртгэлийн дугаар', 'Үйл ажиллагааны чиглэл', 'Утас', 'Факсын дугаар', 'Хаяг']

data = pd.DataFrame(columns=headers)
data
```

```python
url_link  = [f'https://spa.gov.mn/company/{i}' for i in range(3971,5809)]


ua = UserAgent()
fake_ua = {'user-agent': ua.random}
curren_time = time.time()
for lnk in url_link:
    url = lnk

    try:
        response = requests.get(url=url, headers=fake_ua)

        response.encoding = 'utf-8'

        soup = BeautifulSoup(response.text, 'lxml')
        table = soup.find('table', class_='table table-striped table-hovered b-t text-small').select("tr td:last-child")
        row = []
        for i in table:
            if i == '':
                row.append('')
            else:
                row.append(i.text.strip())
        length = len(data)
        data.loc[length] = row
    except Exception as e:
        pass
    
    
print(f'time:{round(time.time() - curren_time, 2)}')
data
```

```python
data.to_csv('registr_company.csv', index=False)

mydata2 = pd.read_csv('registr_company.csv')

```
